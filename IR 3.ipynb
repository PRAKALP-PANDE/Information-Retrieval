```python
import pandas as pd
from pgmpy.models import BayesianNetwork
from pgmpy.estimators import MaximumLikelihoodEstimator
from pgmpy.inference import VariableElimination

```

```python
data = pd.read_csv('/Users/sonavaneonkar/Downloads/Practicals/IR Sem 7/heart_cleveland_upload.csv')
```

```python
model = BayesianNetwork([('age', 'condition'), 
                         ('sex', 'condition'), 
                         ('cp', 'condition'), 
                         ('trestbps', 'condition'), 
                         ('chol', 'condition')])
```

```python
model.fit(data, estimator = MaximumLikelihoodEstimator)
HeartDisease_infer = VariableElimination(model)
```

```python
queries = [
    {'cp': 3},
    {'trestbps': 140},
    {'age': 50},
    {'chol': 220}
]
```

```python
for q in queries:
    result = HeartDisease_infer.query(variables = ['condition'], evidence = q)
    print(result)
```
